```python
import numpy as np 
import pandas as pd
import sklearn as skl
```

```python
songs = pd.read_csv("spotify-2023.csv", encoding="latin-1")
```

```python
# track_name 
# artist(s)_name
songs = songs[['track_name' , 'artist(s)_name' , 'danceability_%' ,'energy_%' ]]
```

```python
songs.isna().sum()
```

```python
songs.dropna()
```

```python
songs["artist(s)_name"] = (
    songs["artist(s)_name"]
    .str.replace(" ", "", regex=False)
    .str.lower()
)
```

```python
songs
```

```python
songs['tags'] = (
    songs['artist(s)_name'].astype(str)
    + " "
    + songs['danceability_%'].astype(str)
    + " "
    + songs['energy_%'].astype(str)
)
```

```python
songs
```

```python
songs = songs[['track_name' , 'tags']]
```

```python
songs
```

```python
songs = songs[['track_name', 'tags']]
songs = songs.dropna().reset_index(drop=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    stop_words='english',
    ngram_range=(1, 2)
)

vectors = tfidf.fit_transform(songs['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(song_name, n=5):
    song_name = song_name.lower()
    
    if song_name not in songs['track_name'].str.lower().values:
        return "Song not found"

    idx = songs[songs['track_name'].str.lower() == song_name].index[0]

    scores = list(enumerate(similarity[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    recommendations = []
    for i in scores[1:n+1]:
        recommendations.append(songs.iloc[i[0]].track_name)

    return recommendations
```

```python
recommend("vampire")
```

```python
import pickle

pickle.dump(similarity, open("similarity.pkl", "wb"))
pickle.dump(songs, open("songs.pkl", "wb"))
```
